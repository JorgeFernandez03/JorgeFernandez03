```python
import random

autos = []


def mostrar_menu():
    print("---- Menú principal ----")
    print("1. Grabar vehículo")
    print("2. Buscar vehículo")
    print("3. Imprimir certificados")
    print("4. Salir")


def grabar_vehiculo():
    print("Ingrese los datos del vehículo:")
    tipo = input("Tipo (Auto, Camioneta, Furgón): ")
    patente = input("Patente: ")
    marca = input("Marca: ")
    modelo = input("Modelo: ")
    color = input("Color: ")
    combustible = input("Combustible (Gasolina o Petróleo): ")
    año = input("Año: ")
    precio = float(input("Precio: "))
    multas_presenta = input("¿Presenta multas? (S/N): ")
    if multas_presenta.lower() == "s":
        multas_monto = float(input("Monto de las multas: "))
        multas_fecha = input("Fecha de las multas: ")
        multas = {"Monto": multas_monto, "Fecha": multas_fecha}
    else:
        multas = {"Monto": 0, "Fecha": ""}
    fecha_registro = input("Fecha de registro del vehículo: ")
    nombre_dueño = input("Nombre del dueño: ")
    rut_dueño = input("RUT del dueño: ")


    if not validar_marca(marca):
        print("La marca ingresada no es válida.")
        return

    if not validar_rut(rut_dueño):
        print("El RUT del dueño ingresado no es válido.")
        return

    if precio < 5000000:
        print("El precio del vehículo debe ser igual o mayor a $5,000,000.")
        return

    vehiculo = {
        "Tipo": tipo,
        "Patente": patente,
        "Marca": marca,
        "Modelo": modelo,
        "Color": color,
        "Combustible": combustible,
        "Año": año,
        "Precio": precio,
        "Multas": multas,
        "Fecha Registro": fecha_registro,
        "Dueño": {
            "Nombre": nombre_dueño,
            "RUT": rut_dueño
        }
    }

    autos.append(vehiculo)
    print("El vehículo ha sido registrado correctamente.")


def buscar_vehiculo():
    patente = input("Ingrese la patente del vehículo: ")
    encontrado = False
    for vehiculo in autos:
        if vehiculo["Patente"] == patente:
            mostrar_informacion(vehiculo)
            encontrado = True
            break
    if not encontrado:
        print("No se encontró un vehículo con la patente ingresada.")


def imprimir_certificado():
    print("Seleccione el certificado a imprimir:")
    print("1. Emisión de contaminantes")
    print("2. Anotaciones vigentes")
    print("3. Multas")
    opcion = int(input("Opción: "))

    if opcion == 1:

        print("Opción no implementada")
    elif opcion == 2:

        print("Opción no implementada")
    elif opcion == 3:
        patente = input("Ingrese la patente del vehículo: ")
        encontrado = False
        for vehiculo in autos:
            if vehiculo["Patente"] == patente:
                imprimir_certificado_multas(vehiculo)
                encontrado = True
                break
        if not encontrado:
            print("No se encontró un vehículo con la patente ingresada.")
    else:
        print("Opción inválida")


def imprimir_certificado_multas(vehiculo):
    folio = str(random.randint(1, 999999999)).zfill(9)
    codigo_verificacion = str(random.randint(11111111, 99999999))

    print(" _______________________")
    print("|servicio de registro                          folio:   " + folio + "|")
    print("|civil e identificacion                       codigo verificacion:|")
    print("|                                                        " + codigo_verificacion + "|")
    print("|_______________________|")
    print("|            CERTIFICADO DE REGISTRO DE MULTAS                    |")
    print("|                 DE TRANSITO NO PAGADAS                          |")
    print("|_______________________|")
    print("|                   Art.24 LEY 18.287                             |")
    print("| vehículo placa patente: ", vehiculo["Patente"], " " * (60 - len(vehiculo["Patente"])), "|")
    print("| Anotaciones de multas de transito no pagadas                    |")
    print("| ID multa   : " + str(random.randint(10000000, 99999999)) +                                                 "    |")
    print("| infracción :  transitar sin dispositivo electrónico             |")
    print("|               por autopista concesionada                        |")
    print("| tribunal   :  primera comisaria Santiago central                |")
    print("| ROL        :  Z" + str(random.randint(10000, 99999)) + "        |")
    print("| fecha infracción   :  00-00-0000                                |")
    print("| fecha sentencia    :  00-00-0000 00:00:00                       |")
    print("| monto multa        :  1,00                                      |")
    print("| arancel            :  2.700,00                                  |")
    print("| tipo moneda multa  :  UTM                                       |")
    print("| multa tipo tag     :  1                                         |")
    print("| fecha informe RMNP :  00-00-0000                                |")
    print("| fecha ingreso RMNP :  00-00-0000 11:22:30                       |")
    print("|                                                                 |")
    print("| DATOS DEL PROPIETARIO                                           |")
    print("| nombre            :  " + vehiculo["Dueño"]["Nombre"] + " " * (50 - len(vehiculo["Dueño"]["Nombre"])) + "|")
    print("| R.U.N.            :  " + vehiculo["Dueño"]["RUT"] + " " * (46 - len(vehiculo["Dueño"]["RUT"])) + "|")
    print("| fecha adquisicion :  " + vehiculo["Fecha Registro"] + " " * (48 - len(vehiculo["Fecha Registro"])) + "|")
    print("|_______________________|")


def mostrar_informacion(vehiculo):
    print("Tipo:", vehiculo["Tipo"])
    print("Patente:", vehiculo["Patente"])
    print("Marca:", vehiculo["Marca"])
    print("Modelo:", vehiculo["Modelo"])
    print("Color:", vehiculo["Color"])
    print("Combustible:", vehiculo["Combustible"])
    print("Año:", vehiculo["Año"])
    print("Precio:", vehiculo["Precio"])
    print("Multas:", vehiculo["Multas"])
    print("Fecha de registro:", vehiculo["Fecha Registro"])
    print("Nombre del dueño:", vehiculo["Dueño"]["Nombre"])
    print("RUT del dueño:", vehiculo["Dueño"]["RUT"])


def validar_patente(patente):
    if len(patente) == 6:
        letras = patente[:2]
        numeros = patente[2:]
        if letras.isalpha() and numeros.isdigit():
            return True
    elif len(patente) == 7:
        letras = patente[:4]
        numeros = patente[4:]
        if letras.isalpha() and numeros.isdigit():
            return True
    return False


def validar_marca(marca):
    return 2 <= len(marca) <= 15


def validar_rut(rut):
    rut_sin_digito = rut[:-2]
    digito_verificador = rut[-1]
    if not rut_sin_digito.isdigit() or not digito_verificador.isdigit():
        return False
    return True


def programa_autoseguro():
    nombre = input("Ingrese su nombre: ")
    apellido = input("Ingrese su apellido: ")
    version = "1.0"

    print("Bienvenido/a,", nombre, apellido)
    print("Programa Auto Seguro - Versión", version)

    while True:
        mostrar_menu()
        opcion = int(input("Seleccione una opción: "))

        if opcion == 1:
            grabar_vehiculo()
        elif opcion == 2:
            buscar_vehiculo()
        elif opcion == 3:
            imprimir_certificado()
        elif opcion == 4:
            print("Gracias por utilizar el programa Auto Seguro. ¡Hasta luego!")
            break
        else:
            print("Opción inválida")
programa_autoseguro()
```
